````python
import matplotlib.pyplot as plt
import numpy as np
samples = 100
frequency1 = 7
frequency2 = 3
````

Sound(t) = A * sin(2Ï€ * f * t)
:Where:

Sound(t) represents the sound at a specific time t.
A is the amplitude of the sound wave.
f is the frequency of the sound wave.
t is the time.
This equation describes a sinusoidal wave representing a sound wave, where the amplitude A determines the maximum displacement of the wave from its equilibrium position, and the frequency f determines the number of cycles of the wave that occur per unit of time.








````python
x = np.arange(samples)
````

````python
x
````

````python
y1 = np.sin(2*np.pi*frequency1*(x/samples))
````

````python
y1.shape
````

````python
y2 = np.sin(2*np.pi*frequency2*(x/samples))
````

````python
y = y1 + y2
````

````python
plt.figure(figsize=(14, 10))
def show_plot(x, y, i):
  plt.subplot(3,1, i)
  plt.stem(x, y, 'r', use_line_collection=True)
  plt.plot(x,y)
  plt.xlabel('time')
  plt.ylabel('amplitude')
  plt.grid()
````

````python
show_plot(x,y1, 1)
show_plot(x, y2, 2)
show_plot(x, y, 3)
````

import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# Calculates the length of the signal x
n = len(x)

# Calculates the sampling period based on the number of samples
T = 1 / samples

# Performs the FFT (Fast Fourier Transform) on the signal y using TensorFlow's signal module.
# This computes the frequency spectrum of the signal.
yf = tf.signal.fft(y)

# Creates the frequency axis using NumPy's linspace function.
# It ranges from 0 to the Nyquist frequency (half of the sampling frequency) with a total of n//2 points.
xf = np.linspace(0.0, 1/(2.0*T), n//2)

# Plots the frequency spectrum.
# It plots the absolute values of the FFT result (magnitude spectrum) up to the Nyquist frequency.
# The factor 2.0/n is applied to normalize the amplitude.
plt.plot(xf, 2.0/n*np.abs(yf[:n//2]))

# Sets the title of the plot to 'spectrum'
plt.title('spectrum')

# Sets the label for the x-axis to 'frequency bins'
plt.xlabel('frequency bins')

# Sets the label for the y-axis to 'Amplitude'
plt.ylabel('Amplitude')

# Displays grid lines on the plot
plt.grid()

# Displays the plot
plt.show()
